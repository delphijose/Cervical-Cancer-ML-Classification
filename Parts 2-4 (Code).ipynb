## Section 4: Plan for Missing Data

This section contains code for filling in missing values. Features with binary indicator data will be filled in with the training data's median. Features with continuous data will be filled in with the training data's mean.

```python
missing_vals_binary = ['Smokes','Hormonal Contraceptives','IUD','STDs','STDs:condylomatosis','STDs:cervical condylomatosis',
'STDs:vaginal condylomatosis','STDs:vulvo-perineal condylomatosis','STDs:syphilis','STDs:pelvic inflammatory disease',
'STDs:genital herpes','STDs:molluscum contagiosum','STDs:AIDS','STDs:HIV','STDs:Hepatitis B','STDs:HPV']

# Calculate medians for the binary indicator columns with missing values (see data dictionary)
medians = train_set[missing_vals_binary].median()

# Fill missing values with the median for each column
train_set[missing_vals_binary] = train_set[missing_vals_binary].fillna(medians)

missing_vals_continous = ['Number of sexual partners','First sexual intercourse','Num of pregnancies',
'Smokes (years)','Smokes (packs/year)','Hormonal Contraceptives (years)','IUD (years)','STDs (number)',
'STDs: Time since first diagnosis','STDs: Time since last diagnosis']

# Calculate means for the continous columns with missing values (see data dictionary)
means = train_set[missing_vals_continous].mean()

# Fill missing values with the mean for each column
train_set[missing_vals_continous] = train_set[missing_vals_continous].fillna(means)

# Fill in missing values in test data with training data medians and means
test_set[missing_vals_binary] = test_set[missing_vals_binary].fillna(medians)
test_set[missing_vals_continous] = test_set[missing_vals_continous].fillna(means)
```

## Section 5: Transformations

This section contains code for transforming the data. Parameters that contain the same value in all rows were removed from the feature space. Continous features were standardized using StandardScaler. Binary indicator features were not standardized.


```python
# Find columns with only one unique value in the original dataset
single_value_cols = [col for col in data.columns if data[col].nunique() == 1]

# Drop these columns from the training and test data

train_set = train_set.drop(columns=single_value_cols)
test_set = test_set.drop(columns=single_value_cols)

# Print removed columns
print("Removed columns with a single unique value:", single_value_cols)
```

```python
# Standardize continous features only

import pandas as pd
from sklearn.preprocessing import StandardScaler

continuous_features = ['Number of sexual partners','First sexual intercourse','Num of pregnancies',
'Smokes (years)','Smokes (packs/year)','Hormonal Contraceptives (years)','IUD (years)','STDs (number)',
'STDs: Time since first diagnosis','STDs: Time since last diagnosis', 'Age', 'STDs: Number of diagnosis']

binary_features = ['Smokes','Hormonal Contraceptives','IUD','STDs','STDs:condylomatosis',
'STDs:vaginal condylomatosis','STDs:vulvo-perineal condylomatosis','STDs:syphilis','STDs:pelvic inflammatory disease',
'STDs:genital herpes','STDs:molluscum contagiosum','STDs:HIV','STDs:Hepatitis B','STDs:HPV',
'Dx:Cancer','Dx:CIN','Dx:HPV','Dx','Hinselmann','Schiller','Citology','Biopsy']

# Initialize the scaler and fit only on train_set continuous features. Transform on train_set and test_set.
scaler = StandardScaler()
train_continuous_scaled = scaler.fit_transform(train_set[continuous_features])
test_continuous_scaled = scaler.transform(test_set[continuous_features])

# Combine scaled continuous and binary features for both datasets
train_standardized = pd.concat([
    pd.DataFrame(train_continuous_scaled, columns=continuous_features).reset_index(drop=True),
    train_set[binary_features].reset_index(drop=True)
], axis=1)

test_standardized = pd.concat([
    pd.DataFrame(test_continuous_scaled, columns=continuous_features).reset_index(drop=True),
    test_set[binary_features].reset_index(drop=True)
], axis=1)
```

## Section 6: Additional Dimensionality Reduction

```python
# Split features and target in the training set
X_train = train_standardized.drop(columns=['Biopsy'])
y_train = train_standardized['Biopsy']

# Split features and target in the test set
X_test = test_standardized.drop(columns=['Biopsy'])
y_test = test_standardized['Biopsy']
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

# Function to plot explained variance ratio for PCA
def plot_pca_variance(X_train):
    pca = PCA()
    pca.fit(X_train)

    # Plot the explained variance ratio for each component
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
    plt.title('Cumulative Explained Variance by PCA Components')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.grid(True)
    plt.show()

plot_pca_variance(X_train)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# With LDA for dimensionality reduction
lda = LinearDiscriminantAnalysis(n_components=1) # Picked 1 component since number of classes = 2
X_train_lda = lda.fit_transform(X_train, y_train)
X_test_lda = lda.transform(X_test)

# With PCA for dimensionality reduction
pca = PCA(n_components=7)  # Picked 7 components based on variance graph above
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

## Section 7: K Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import GridSearchCV

# Function to train and evaluate KMeans model
def apply_kmeans(X_train, y_train, X_test, y_test, n_clusters=2, init='k-means++', n_init=10, max_iter=300, tol=1e-4, algorithm='auto'):
    # Initialize KMeans with the given hyperparameters
    kmeans = KMeans(
        n_clusters=n_clusters,
        init=init,
        n_init=n_init,
        max_iter=max_iter,
        tol=tol,
        algorithm=algorithm,
        random_state=0
    )
    
    # Fit the model and get predicted clusters
    clusters_train = kmeans.fit_predict(X_train)

    # Map clusters to majority labels
    cluster_mapping = {}
    for cluster in range(n_clusters):
        majority_label = y_train[clusters_train == cluster].mode()[0]
        cluster_mapping[cluster] = majority_label

    # Map training clusters to predicted labels
    y_train_pred = np.array([cluster_mapping[cluster] for cluster in clusters_train])

    # Predict clusters for the test set
    clusters_test = kmeans.predict(X_test)
    y_test_pred = np.array([cluster_mapping[cluster] for cluster in clusters_test])

    # Evaluate on training data
    train_conf_matrix = confusion_matrix(y_train, y_train_pred)
    train_accuracy = accuracy_score(y_train, y_train_pred)

    # Evaluate on test data
    test_conf_matrix = confusion_matrix(y_test, y_test_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)

    return {
        "train_conf_matrix": train_conf_matrix,
        "train_accuracy": train_accuracy,
        "test_conf_matrix": test_conf_matrix,
        "test_accuracy": test_accuracy,
        "clusters_train": clusters_train,
        "clusters_test": clusters_test,
    }

# Hyperparameter tuning with GridSearchCV
def tune_kmeans(X_train, y_train):
    param_grid = {
        'n_clusters': [2, 3, 4, 5, 6],  # Number of clusters
        'init': ['k-means++', 'random'],  # Initialization method
        'n_init': [5, 10, 15],  # Number of initializations
        'max_iter': [300, 500],  # Maximum number of iterations
        'tol': [1e-4, 1e-3],  # Tolerance
        'algorithm': ['lloyd', 'elkan']  # Algorithm choice
    }
    kmeans = KMeans(random_state=0)

    # GridSearchCV with KMeans
    grid_search = GridSearchCV(kmeans, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    # Best hyperparameters
    print(f"Best hyperparameters: {grid_search.best_params_}")
    return grid_search.best_params_

# Function to plot clusters in 2D
def plot_clusters(X, y_km, n_clusters, dim1=0, dim2=1):
    color_list = ['lightgreen', 'orange']
    plt.figure(figsize=(8, 6))
    for i in range(n_clusters):
        plt.scatter(X[y_km == i, dim1], X[y_km == i, dim2], s=50, c=color_list[i], label=f'Cluster {i}')
    plt.title("K Means Clustering with PCA")
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

# Find the best hyperparameters via cross-validation
best_params = tune_kmeans(X_train, y_train)

# Apply KMeans with the best hyperparameters found via cross-validation
results_no_reduction = apply_kmeans(X_train, y_train, X_test, y_test, 
                                    n_clusters=best_params['n_clusters'],
                                    init=best_params['init'],
                                    n_init=best_params['n_init'],
                                    max_iter=best_params['max_iter'],
                                    tol=best_params['tol'],
                                    algorithm=best_params['algorithm'])

# Apply KMeans with LDA-reduced data and best hyperparameters
results_with_lda = apply_kmeans(X_train_lda, y_train, X_test_lda, y_test, 
                                n_clusters=best_params['n_clusters'],
                                init=best_params['init'],
                                n_init=best_params['n_init'],
                                max_iter=best_params['max_iter'],
                                tol=best_params['tol'],
                                algorithm=best_params['algorithm'])

# Apply KMeans with PCA-reduced data and best hyperparameters
results_with_pca = apply_kmeans(X_train_pca, y_train, X_test_pca, y_test, 
                                n_clusters=best_params['n_clusters'],
                                init=best_params['init'],
                                n_init=best_params['n_init'],
                                max_iter=best_params['max_iter'],
                                tol=best_params['tol'],
                                algorithm=best_params['algorithm'])

print("\nResults without dimensionality reduction:")
print("Training Confusion Matrix:")
print(results_no_reduction["train_conf_matrix"])
print(f"Training Accuracy: {results_no_reduction['train_accuracy']:.4f}")
print("\nTest Confusion Matrix:")
print(results_no_reduction["test_conf_matrix"])
print(f"Test Accuracy: {results_no_reduction['test_accuracy']:.4f}")

print("\nResults with LDA dimensionality reduction:")
print("Training Confusion Matrix:")
print(results_with_lda["train_conf_matrix"])
print(f"Training Accuracy: {results_with_lda['train_accuracy']:.4f}")
print("\nTest Confusion Matrix:")
print(results_with_lda["test_conf_matrix"])
print(f"Test Accuracy: {results_with_lda['test_accuracy']:.4f}")

print("\nResults with PCA dimensionality reduction:")
print("Training Confusion Matrix:")
print(results_with_pca["train_conf_matrix"])
print(f"Training Accuracy: {results_with_pca['train_accuracy']:.4f}")
print("\nTest Confusion Matrix:")
print(results_with_pca["test_conf_matrix"])
print(f"Test Accuracy: {results_with_pca['test_accuracy']:.4f}")

plot_clusters(X_train_pca, results_with_pca["clusters_train"], best_params['n_clusters'], dim1=0, dim2=1)
```

## Section 8: Linear Kernel SVM

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report

# Function to train and evaluate SVM
def train_and_evaluate_svm(X_train, y_train, X_test, y_test, svm_model=None):
    # Train the model
    svm_model.fit(X_train, y_train)
    
    # Predict on training data
    y_train_pred = svm_model.predict(X_train)
    
    # Predict on test data
    y_test_pred = svm_model.predict(X_test)
    
    # Evaluate on training data
    train_conf_matrix = confusion_matrix(y_train, y_train_pred)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_class_report = classification_report(y_train, y_train_pred)
    
    # Evaluate on test data
    test_conf_matrix = confusion_matrix(y_test, y_test_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_class_report = classification_report(y_test, y_test_pred)
    
    return {
        "train_conf_matrix": train_conf_matrix,
        "train_accuracy": train_accuracy,
        "train_class_report": train_class_report,
        "test_conf_matrix": test_conf_matrix,
        "test_accuracy": test_accuracy,
        "test_class_report": test_class_report,
    }

# Hyperparameter tuning with GridSearchCV
def tune_svm(X_train, y_train):
    param_grid = {
        'C': [0.1, 1, 10, 100],  # Regularization parameter
        'max_iter': [1000, 5000, 10000],  # Maximum iterations
        'tol': [1e-3, 1e-4, 1e-5],  # Tolerance for stopping criteria
    }

    svm = LinearSVC(class_weight='balanced', random_state=42, dual=False)

    # GridSearchCV with SVM
    grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Best hyperparameters
    print(f"Best hyperparameters: {grid_search.best_params_}")
    return grid_search.best_params_, grid_search.best_estimator_

# Function to plot decision boundary
def plot_decision_boundary(X, y, model, title="SVM Decision Boundary"):
    h = .02 
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.75, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', s=50, cmap=plt.cm.coolwarm)
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.title(title)
    plt.show()

# Find the best hyperparameters via cross-validation
best_params, best_svm_model = tune_svm(X_train, y_train)

# Apply the best SVM model without dimensionality reduction
results_no_reduction = train_and_evaluate_svm(X_train, y_train, X_test, y_test, best_svm_model)

# Apply the best SVM model with LDA-reduced data
results_with_lda = train_and_evaluate_svm(X_train_lda, y_train, X_test_lda, y_test, best_svm_model)

# Apply the best SVM model with PCA-reduced data
results_with_pca = train_and_evaluate_svm(X_train_pca, y_train, X_test_pca, y_test, best_svm_model)

# Plot decision boundary for PCA data with 2 components
pca2 = PCA(n_components=2) 
X_train_pca2 = pca2.fit_transform(X_train)
X_test_pca2 = pca2.transform(X_test)
results_with_pca2 = train_and_evaluate_svm(X_train_pca2, y_train, X_test_pca2, y_test, best_svm_model)
plot_decision_boundary(X_train_pca2, y_train, best_svm_model, title="Linear Kernel SVM Decision Boundary with PCA")

print("\nResults without dimensionality reduction:")
print("Training Confusion Matrix:")
print(results_no_reduction["train_conf_matrix"])
print(f"Training Accuracy: {results_no_reduction['train_accuracy']:.4f}")
print("\nTraining Classification Report:")
print(results_no_reduction["train_class_report"])
print("\nTest Confusion Matrix:")
print(results_no_reduction["test_conf_matrix"])
print(f"Test Accuracy: {results_no_reduction['test_accuracy']:.4f}")
print("\nTest Classification Report:")
print(results_no_reduction["test_class_report"])

print("\nResults with LDA dimensionality reduction:")
print("Training Confusion Matrix:")
print(results_with_lda["train_conf_matrix"])
print(f"Training Accuracy: {results_with_lda['train_accuracy']:.4f}")
print("\nTraining Classification Report:")
print(results_with_lda["train_class_report"])
print("\nTest Confusion Matrix:")
print(results_with_lda["test_conf_matrix"])
print(f"Test Accuracy: {results_with_lda['test_accuracy']:.4f}")
print("\nTest Classification Report:")
print(results_with_lda["test_class_report"])

print("\nResults with PCA dimensionality reduction:")
print("Training Confusion Matrix:")
print(results_with_pca["train_conf_matrix"])
print(f"Training Accuracy: {results_with_pca['train_accuracy']:.4f}")
print("\nTraining Classification Report:")
print(results_with_pca["train_class_report"])
print("\nTest Confusion Matrix:")
print(results_with_pca["test_conf_matrix"])
print(f"Test Accuracy: {results_with_pca['test_accuracy']:.4f}")
print("\nTest Classification Report:")
print(results_with_pca["test_class_report"])
```

```python
from sklearn.linear_model import LogisticRegression

# Function to train and evaluate Logistic Regression
def train_and_evaluate_logistic_regression(X_train, y_train, X_test, y_test, logreg_model=None):
    # Train the model
    logreg_model.fit(X_train, y_train)
    
    # Predict on training data
    y_train_pred = logreg_model.predict(X_train)
    
    # Predict on test data
    y_test_pred = logreg_model.predict(X_test)
    
    # Evaluate on training data
    train_conf_matrix = confusion_matrix(y_train, y_train_pred)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_class_report = classification_report(y_train, y_train_pred)
    
    # Evaluate on test data
    test_conf_matrix = confusion_matrix(y_test, y_test_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_class_report = classification_report(y_test, y_test_pred)
    
    return {
        "train_conf_matrix": train_conf_matrix,
        "train_accuracy": train_accuracy,
        "train_class_report": train_class_report,
        "test_conf_matrix": test_conf_matrix,
        "test_accuracy": test_accuracy,
        "test_class_report": test_class_report,
    }

# Hyperparameter tuning with GridSearchCV for Logistic Regression
def tune_logistic_regression(X_train, y_train):
    param_grid = {
        'C': [0.1, 1, 10, 100],  # Regularization parameter
        'max_iter': [1000, 5000, 10000],  # Max iterations for convergence
        'tol': [1e-3, 1e-4, 1e-5]  # Tolerance for stopping criteria
    }

    logreg = LogisticRegression(class_weight='balanced', random_state=42)

    # GridSearchCV for logistic regression
    grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Best hyperparameters
    print(f"Best hyperparameters: {grid_search.best_params_}")
    return grid_search.best_params_, grid_search.best_estimator_

# Find the best hyperparameters via cross-validation
best_params, best_logreg_model = tune_logistic_regression(X_train, y_train)

# Apply the best Logistic Regression model without dimensionality reduction
results_no_reduction = train_and_evaluate_logistic_regression(X_train, y_train, X_test, y_test, best_logreg_model)

# Apply the best Logistic Regression model with LDA-reduced data
results_with_lda = train_and_evaluate_logistic_regression(X_train_lda, y_train, X_test_lda, y_test, best_logreg_model)

# Apply the best Logistic Regression model with PCA-reduced data
results_with_pca = train_and_evaluate_logistic_regression(X_train_pca, y_train, X_test_pca, y_test, best_logreg_model)

results_with_pca2 = train_and_evaluate_logistic_regression(X_train_pca2, y_train, X_test_pca2, y_test, best_logreg_model)

plot_decision_boundary(X_train_pca2, y_train, best_logreg_model, title="Logistic Regression with PCA")

print("\nResults without dimensionality reduction:")
print("Training Confusion Matrix:")
print(results_no_reduction["train_conf_matrix"])
print(f"Training Accuracy: {results_no_reduction['train_accuracy']:.4f}")
print("\nTraining Classification Report:")
print(results_no_reduction["train_class_report"])
print("\nTest Confusion Matrix:")
print(results_no_reduction["test_conf_matrix"])
print(f"Test Accuracy: {results_no_reduction['test_accuracy']:.4f}")
print("\nTest Classification Report:")
print(results_no_reduction["test_class_report"])

print("\nResults with LDA dimensionality reduction:")
print("Training Confusion Matrix:")
print(results_with_lda["train_conf_matrix"])
print(f"Training Accuracy: {results_with_lda['train_accuracy']:.4f}")
print("\nTraining Classification Report:")
print(results_with_lda["train_class_report"])
print("\nTest Confusion Matrix:")
print(results_with_lda["test_conf_matrix"])
print(f"Test Accuracy: {results_with_lda['test_accuracy']:.4f}")
print("\nTest Classification Report:")
print(results_with_lda["test_class_report"])

print("\nResults with PCA dimensionality reduction:")
print("Training Confusion Matrix:")
print(results_with_pca["train_conf_matrix"])
print(f"Training Accuracy: {results_with_pca['train_accuracy']:.4f}")
print("\nTraining Classification Report:")
print(results_with_pca["train_class_report"])
print("\nTest Confusion Matrix:")
print(results_with_pca["test_conf_matrix"])
print(f"Test Accuracy: {results_with_pca['test_accuracy']:.4f}")
print("\nTest Classification Report:")
print(results_with_pca["test_class_report"])
```
